# 01 Extraction and Initial Audit

**Purpose:** Load the raw retail sales dataset, inspect its structure, print basic metadata (shape, dtypes), and document initial observations to inform the cleaning strategy.

```python
import pandas as pd
from pathlib import Path

# Define paths relative to project root
PROJECT_ROOT = Path.cwd().parent
RAW_DATA_PATH = PROJECT_ROOT / 'data' / 'raw' / 'retail_store_sales.csv'
```

### Load the Data
Loading the CSV file using pandas and printing its shape.

```python
df = pd.read_csv(RAW_DATA_PATH)
print(f"Dataset Shape: {df.shape}")
```

### Inspect Data Types and Missing Values
Using `.info()` to understand column data types and initial null value counts.

```python
df.info()
```

### View Sample Data
Looking at the first few rows to understand formatting.

```python
df.head()
```

### Initial Observations for Data Cleaning (`02_cleaning.ipynb`)
- **Column Names**: They contain spaces and mixed casing; they need standardization to `snake_case`.
- **Discount Applied**: Contains boolean/string variations that must be mapped to 0/1 integers.
- **Transaction Date**: Currently an object/string and needs parsing to datetime.
- **Missing Values**: There are missing values in numerical columns (`Price Per Unit`, `Total Spent`) that require imputation or dropping.
- **Categorical Variations**: String columns like Category might need standardizing to Title Case.

### Update Data Dictionary
Saving the initial snapshot of columns to `docs/data_dictionary.md`.

```python
dictionary_path = PROJECT_ROOT / 'docs' / 'data_dictionary.md'
dictionary_path.parent.mkdir(parents=True, exist_ok=True)

with open(dictionary_path, 'w') as f:
    f.write("# Data Dictionary: Retail Store Sales\n\n")
    f.write("| Column | Original Dtype | Description |\n")
    f.write("|---|---|---|\n")
    for col, dtype in zip(df.columns, df.dtypes):
        f.write(f"| {col} | {dtype} | Original raw column |\n")

print("Data dictionary snapshot saved!")
```
